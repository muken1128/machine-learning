```python
import numpy as np
import pandas as pd

# カリフォルニアの住宅価格データ
from sklearn.datasets import fetch_california_housing
# データ分割
from sklearn.model_selection import train_test_split
# 重回帰分析
from sklearn.linear_model import LinearRegression
# 評価指標：決定係数
from sklearn.metrics import r2_score
```

```python
ca_data = fetch_california_housing()

# 説明変数
x = ca_data.data
# 目的変数
y = ca_data.target

# データフレームへ格納
df_x = pd.DataFrame(x, columns=ca_data.feature_names)
df_y = pd.DataFrame(y, columns=["price"])
# 説明変数と目的変数の内容を表示
display(df_x.head())
display(df_y.head())
```

```python
# test_size：テストデータの割合
# random_state：乱数シード値
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# 重回帰モデル
rl_model = LinearRegression()
```

```python
# 学習
rl_model.fit(x_train,y_train)

# テスト
y_pred_test = rl_model.predict(x_test)

# 決定係数
r2_test = r2_score(y_test, y_pred_test)
# テスト結果の表示
print("R2スコア（テストデータ）:", r2_test)
```
